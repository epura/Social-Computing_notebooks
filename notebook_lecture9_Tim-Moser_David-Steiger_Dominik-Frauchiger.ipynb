### 1. Identifying the country of residence

Based on the email address, we can infer different properties about the user. One example is their country of residence. For example, an email like “1315ae6229444367968a943a219f38def9a8112d@vpn-251-169.epfl.ch” will probably correspond to a user located in Switzerland, while an email of the form “59bd0a3ff43b32849b319e645d4798d8a5d1e889@philipphauer.de” will probably correspond to a user located in Germany. 

1. Write a script that classifies the email addresses based on the country of residence.

```python
# No output in the console, outputs to file.
# This takes about 8 minutes.
import requests, os, io # io for unicode
from bs4 import BeautifulSoup
#from tqdm import tqdm # library for a progress bar

# saves the table from wikipedia
# only downloads if file not exists
if not os.path.isfile('country-code-table.html'):
    source = requests.get('https://en.wikipedia.org/wiki/List_of_Internet_top-level_domains').text
    soup = BeautifulSoup(source, 'html.parser')
    with io.open('country-code-table.html', "w", encoding='utf-8') as html:
        html.write(soup.select_one('table.wikitable:nth-child(17)').prettify())


# opens csv and table
with io.open('hashedEmailAddressesGitHubCommits.csv', 'r', encoding='utf-8') as csv:
    with io.open('country-code-table.html', 'r', encoding='utf-8') as table:
        with io.open('task1-output.csv', 'w', encoding='utf-8') as output:
            soup = BeautifulSoup(table, 'html.parser')
            already_parsed = dict()

            for line in csv: # for line in tqdm(csv):
                # remove newline
                line = line.rstrip()
                output.write(line)
                # gets only the top level domain
                split_tld = line.split(".")
                tld = "." + split_tld[len(split_tld)-1]
                
                # Saving the found values in a dictionary saves A LOT of time
                # Without this we had first 30 minutes for 10mb
                # Now it completes the whole file in 8 minutes.
                country_name = ""
                if (tld in already_parsed):
                    country_name = already_parsed[tld]
                else:
                    # gets the table entry of the tld
                    country_name = soup.find('a', title=tld)
                    if (country_name):
                        sibling = country_name.parent.find_next_siblings()
                        if (0 == len(sibling)) :
                            country_name = "error"
                        else:
                            # Removes all whitespaces and add one space between words
                            country_name = " ".join(sibling[0].text.split())
                    else:
                        country_name = "Not a country-code"
                    # saves parsed value in dictionary for later use
                    already_parsed[tld] = country_name
                # output to file, should open in excel
                output.write(";" + country_name + "\n")
```
